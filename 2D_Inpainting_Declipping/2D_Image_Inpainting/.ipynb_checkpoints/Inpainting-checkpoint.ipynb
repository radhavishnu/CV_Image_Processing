```python
def lsinpaint(u0,D):
    I=u0
    #I[I==0] = np.NaN
    he, wi=I.shape;

    x1=[];

    for i in range(he):
        y=(I[i,:]).T;

        if sum(np.isnan(y))>0:
            N = len(y);
            n = np.array([i for i in range(N)]);
            e = np.ones((N, 1));
            D = sc.sparse.spdiags(np.hstack((e, -2*e, e)).T, [0, 1, 2], N, N);
            k = np.isfinite(y);                    # k : logical vector, indexes known values

            S = np.eye(N);
            S[np.logical_not(k), :] = 0;                      # S : sampling matrix

            Sc = np.eye(N);                      # Sc : complement of S
            Sc[k, :] = 0;

            L = sum(np.logical_not(k)) ;                       # L : number of missing values
            
            v =  np.linalg.solve(-(Sc @ (D.T @ D) @ Sc.T), ( Sc @ D.T @ D @ S.T @ y[k]))   # v : estimated samples   

            x = np.zeros((N,1));
            x[k] = y[k];
            x[np.logical_not(k)] = v;
        else:
            x=y;
        try:
            x1=np.vstack((x1,x.T));
        except:
            x1 = x.T
        

    x2=[];
    for i in range(wi):
        y=(I[:,i]);

        if sum(np.isnan(y))>0:
            N = len(y);
            n = np.array([i for i in range(N)]);
            e = np.ones((N, 1));
            D = sc.sparse.spdiags(np.hstack((e, -2*e, e)).T, [0, 1, 2], N, N);
            k = np.isfinite(y);                    # k : logical vector, indexes known values

            S = np.eye(N);
            S[np.logical_not(k), :] = 0;                      # S : sampling matrix

            Sc = np.eye(N);                      # Sc : complement of S
            Sc[k, :] = 0;

            L = sum(np.logical_not(k)) ;                       # L : number of missing values

            v =  np.linalg.solve(-(Sc @ (D.T @ D) @ Sc.T), ( Sc @ D.T @ D @ S.T @ y[k]))   # v : estimated samples   

            x = np.zeros((N,1));
            x[k] = y[k];
            x[np.logical_not(k)] = v;
        else:
            x=y;
        try:
            x2=np.hstack((x2, x));
        except:
            x2 = x

    x=(x1+x2)/2;
          
    return x
```
