```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import spdiags
from scipy import sparse
from scipy.linalg import lstsq
import cv2
```

```python
def LS_DEN(img,lam):
    y=img
    M=np.size(img,0)
    N=np.size(img,1)
    e = np.ones([M,1],dtype=int)
    e1 = np.ones([N,1],dtype=int)
    D = spdiags(np.transpose(np.column_stack((e,-2*e,e))),np.linspace(0,2,3),M,M)
    D1 = spdiags(np.transpose(np.column_stack((e1,-2*e1,e1))),np.linspace(0,2,3),N,N)
    F =sparse.eye(M).toarray() + lam*(np.transpose(D)@D)
    F1 =sparse.eye(N).toarray() + lam*(np.transpose(D1)@D1)
    x,res,rnk,s=lstsq(F,y)
    return x;
    
```
